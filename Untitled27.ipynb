```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
```

```python
x,y=load_breast_cancer(return_X_y=True)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
classifier=DecisionTreeClassifier()
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn import tree
plt.figure(figsize=(20,20))
tree.plot_tree(classifier,filled=True)
```

```python
path = classifier.cost_complexity_pruning_path(x_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
```

```python
ccp_alphas,impurities=path.ccp_alphas,path.impurities
```

```python
ccp_alphas
```

```python
impurities
```

```python
clfs=[]

for ccp_alpha in  ccp_alphas:
  classifier=DecisionTreeClassifier(random_state=0,ccp_alpha=ccp_alpha)
  classifier.fit(x_train,y_train)
  clfs.append(classifier)
```

```python
clfs
```

```python
clfs[-1].tree_.node_count
```

```python
ccp_alphas[-1]
```

```python
train_scores=[classifier.score(x_train,y_train) for clf in clfs]
test_score=[classifier.score(x_test,y_test) for clf in clfs]
```

```python

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker='o', label="train",
        drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker='o', label="test",
        drawstyle="steps-post")
ax.legend()
plt.show()
```
